```python
data1 = pd.read_csv("C:/Users/samba/OneDrive/Desktop/TAMU CORPUS/Driscoll/sarima/datasets/CC_2021.csv")
data2 = data1
```

```python
data2 = pd.DataFrame(data2)
```

```python
data2
```

```python
data2['Date'] = pd.to_datetime(data2['Date'])
data2.index = data2['Date']
```

```python
data2.drop(['Date'], axis = 1,inplace = True)
```

```python
from sklearn.model_selection import train_test_split
dataset_train, dataset_test = train_test_split(data2, test_size=0.095, shuffle=False)
```

```python
training_set = dataset_train.values
training_set
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
X_train = []
y_train = []
for i in range(1, len(training_set_scaled)):
    X_train.append(training_set_scaled[i-1:i, 0])
    y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
print (X_train)
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense
```

```python
#stacked LSTM
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.1))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(units=50))
model.add(Dropout(0.1))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(X_train,y_train,epochs=50,batch_size=4)
```

```python
real_average_eg = dataset_test.values
```

```python
real_average_eg
```

```python
dataset_test
```

```python
dataset_total = pd.concat((dataset_train['CC'], dataset_test['CC']), axis = 0)
```

```python
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 1:].values
```

```python
inputs = inputs.reshape(-1,1)
```

```python
inputs = sc.transform(inputs)
```

```python
X_test = []
for i in range(1, len(dataset_test)+1):
    X_test.append(inputs[i-1:i, 0])
```

```python
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_average_eg = model.predict(X_test)
predicted_average_eg = sc.inverse_transform(predicted_average_eg)
```
